```python
import pandas as pd
```

```python
import numpy as np
```

Importar valores del CSV

```python
df = pd.read_csv("personas_sucio.csv")
```

```python
df = df.copy()
```

EJERCICIOS

- No haya valores nulos

```python
df.fillna("YA NO HAY NULOS")
```

- Todos los nombres tengan la primera letra en mayúscula y el resto minúsculas

```python
df["nombre"] = df["nombre"].str.strip().str.title()
```

```python
df
```

- Todas las edades estén en formato númerico (las que sean 0 o NaN, sustituirlas por el valor medio de la columna EDAD).

```python
df["edad"] = pd.to_numeric(df["edad"], errors="coerce")
```

```python
df
```

```python
df["edad"] = df["edad"].replace(0, np.nan)
```

```python
df
```

```python
df["edad"] = df["edad"].fillna(df["edad"].mean())
```

```python
df
```

- Todas las ciudades estén escritas en mayúsculas

```python
df["ciudad"] = df["ciudad"].str.upper()
```

```python
df
```

- Todos los emails tengan el formato correcto (texto@otrotexto.com)

```python
df = df[df["email"].str.contains(r"^[\w\.-]+@[\w\.-]+\.\w+$", na=False)]
```

```python
df
```

- Todas las fechas estén en formato DD/MM/AAAA.

```python
df["fecha_registro"] = pd.to_datetime(df["fecha_registro"], errors="coerce")
```

```python
df
```

```python
df["fecha_registro"] = df["fecha_registro"].dt.strftime('%d/%m/%Y')
```

```python
df = df.drop(columns=["fecha"])
```
